```python
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
uploaded=files.upload()
data=pd.read_csv('train_weekthree.csv')
```

```python
''' 5. Use Sampling with replacement to increase the data size by 15 percent.
6. Use 15% of the increased data set as the test data. '''

X = data.drop('Class variable (0 or 1)', axis=1)
y = data['Class variable (0 or 1)']

from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
'''
1. Train a neural network on tensorflow for the given data '''

import tensorflow as tf
from tensorflow.keras import layers, models

# Define your neural network model
model = models.Sequential()
model.add(layers.Dense(128, activation='sigmoid', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))  # For binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
test_loss, test_acc_nn = model.evaluate(X_test, y_test)
print('Test accuracy: ',test_acc_nn)
```

```python
'''2. Train a decision tree with sklearn for the given data '''

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# Create the decision tree model
model = DecisionTreeClassifier()

# Train the model
model.fit(X_train, y_train)

# Evaluate the model
test_acc_dt=model.score(X_test, y_test)
print('Test accuracy: ',test_acc_dt)
```

```python
''' 3. Make predictions using the ensemble (average) of the two datasets '''

# Split the data into two parts
X1, X2, y1, y2 = train_test_split(X, y, test_size=0.5, random_state=42)

# Split the data into training and testing sets
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.15, random_state=42)

# Split the data into training and testing sets
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.15, random_state=42)
```

```python
# Model 1 - Train on the first split
model1 = models.Sequential()
model1.add(layers.Dense(64, activation='relu', input_shape=(X1.shape[1],)))
model1.add(layers.Dense(32, activation='relu'))
model1.add(layers.Dense(1, activation='sigmoid'))  # For binary classification
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model1.fit(X1, y1, epochs=10, batch_size=32, validation_split=0.2)

# Model 2 - Train on the second split
model2 = models.Sequential()
model2.add(layers.Dense(64, activation='relu', input_shape=(X2.shape[1],)))
model2.add(layers.Dense(32, activation='relu'))
model2.add(layers.Dense(1, activation='sigmoid'))  # For binary classification
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model2.fit(X2, y2, epochs=10, batch_size=32, validation_split=0.2)

# Make predictions on the entire dataset
predictions1 = model1.predict(X_test)
predictions2 = model2.predict(X_test)

# Average the predictions
ensemble_predictions = (predictions1 + predictions2) / 2
```

```python
from sklearn.metrics import accuracy_score

# Convert the ensemble predictions to binary values (0 or 1) based on a threshold (e.g., 0.5)
ensemble_binary_predictions = (ensemble_predictions > 0.5).astype(int)

# Calculate ensemble accuracy
ensemble_accuracy = accuracy_score(y_test, ensemble_binary_predictions)
print('Ensemble Accuracy: ',ensemble_accuracy)

print('Test accuracy DT: ',test_acc_dt)
print('Test accuracy NN: ',test_acc_nn)
```
